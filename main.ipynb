```python
import numpy as np
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.model_selection import cross_val_score, KFold
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler

def calculate_rmse(y_true, y_pred):
    return np.sqrt(mean_squared_error(y_true, y_pred))

model_results = []
```

```python
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
%pip install xgboost
```

```python
from sklearn.linear_model import ElasticNet, Ridge, Lasso
from sklearn.ensemble import GradientBoostingRegressor, RandomForestRegressor
from sklearn.svm import SVR
from sklearn.tree import DecisionTreeRegressor
from lightgbm import LGBMRegressor
import xgboost as xgb

models = [
    ('Linear Ridge', Ridge(alpha=1.0, random_state=42)),
    ('Lasso', Lasso(alpha=0.01, random_state=42)),
    ('ElasticNet', ElasticNet(alpha=0.01, l1_ratio=0.5, random_state=42)),
    ('Decision Tree', DecisionTreeRegressor(max_depth=10, random_state=42)),
    ('Random Forest', RandomForestRegressor(n_estimators=100, max_depth=15, random_state=42)),
    ('Gradient Boosting', GradientBoostingRegressor(n_estimators=100, learning_rate=0.1, random_state=42)),
    ('LightGBM', LGBMRegressor(n_estimators=100, max_depth=15, learning_rate=0.1, random_state=42)),
    ('XGBoost', xgb.XGBRegressor(n_estimators=100, max_depth=5, learning_rate=0.1, random_state=42)),
    ('SVR (RBF)', SVR(kernel='rbf', C=10, epsilon=0.1, gamma='scale'))
]

results = []
names = []
kfold = KFold(n_splits=5, shuffle=True, random_state=42)

for name, model in models:
    try:
        print(f"Training {name}...")
        model.fit(X_train_scaled, y_train)
        y_pred = model.predict(X_test_scaled)
        rmse_score = calculate_rmse(y_test, y_pred)
        r2 = r2_score(y_test, y_pred)
        
        cv_scores = cross_val_score(model, X_train_scaled, y_train, cv=kfold, 
                                  scoring='neg_root_mean_squared_error')
        cv_rmse = -cv_scores.mean()
        
        model_results.append({
            'name': name,
            'model': model,
            'rmse': rmse_score,
            'cv_rmse': cv_rmse,
            'r2': r2
        })
        
        print(f"{name}: RMSE = {rmse_score:.4f}, CV RMSE = {cv_rmse:.4f}, R² = {r2:.4f}\n")
    except Exception as e:
        print(f"Error with {name}: {e}\n")

results_df = pd.DataFrame(model_results).sort_values(by='rmse')
results_df
```

```python
plt.figure(figsize=(12, 6))
plt.barh(results_df['name'], results_df['rmse'], color='skyblue')
plt.xlabel('RMSE (lower is better)')
plt.ylabel('Model')
plt.title('Model Performance Comparison')
plt.grid(axis='x', linestyle='--', alpha=0.6)

for i, v in enumerate(results_df['rmse']):
    plt.text(v + 0.02, i, f"{v:.4f}", va='center')

plt.tight_layout()
plt.show()
```

## Best Model Selection

Let's select the best performing model based on RMSE and use advanced hyperparameter tuning to further improve it.

```python
best_model_info = results_df.iloc[0]
print(f"Best model: {best_model_info['name']} with RMSE = {best_model_info['rmse']:.4f}")

best_model_type = best_model_info['name']

from sklearn.model_selection import RandomizedSearchCV

if 'LightGBM' in best_model_type:
    model_to_tune = LGBMRegressor(random_state=42)
    param_dist = {
        'n_estimators': np.arange(50, 500, 50),
        'max_depth': np.arange(5, 30, 5),
        'learning_rate': [0.001, 0.01, 0.05, 0.1, 0.2],
        'subsample': [0.6, 0.7, 0.8, 0.9, 1.0],
        'colsample_bytree': [0.6, 0.7, 0.8, 0.9, 1.0],
        'reg_alpha': [0, 0.1, 0.5, 1.0],
        'reg_lambda': [0, 0.1, 0.5, 1.0],
        'num_leaves': [31, 63, 127, 255]
    }
elif 'XGBoost' in best_model_type:
    model_to_tune = xgb.XGBRegressor(random_state=42)
    param_dist = {
        'n_estimators': np.arange(50, 500, 50),
        'max_depth': np.arange(3, 10),
        'learning_rate': [0.001, 0.01, 0.05, 0.1, 0.2],
        'subsample': [0.6, 0.7, 0.8, 0.9, 1.0],
        'colsample_bytree': [0.6, 0.7, 0.8, 0.9, 1.0],
        'gamma': [0, 0.1, 0.2, 0.3, 0.4],
        'reg_alpha': [0, 0.1, 0.5, 1.0],
        'reg_lambda': [0, 0.1, 0.5, 1.0]
    }
elif 'Gradient Boosting' in best_model_type:
    model_to_tune = GradientBoostingRegressor(random_state=42)
    param_dist = {
        'n_estimators': np.arange(50, 500, 50),
        'max_depth': np.arange(3, 10),
        'learning_rate': [0.001, 0.01, 0.05, 0.1, 0.2],
        'subsample': [0.6, 0.7, 0.8, 0.9, 1.0],
        'min_samples_split': [2, 5, 10],
        'min_samples_leaf': [1, 2, 4]
    }
elif 'Random Forest' in best_model_type:
    model_to_tune = RandomForestRegressor(random_state=42)
    param_dist = {
        'n_estimators': np.arange(50, 500, 50),
        'max_depth': [None] + list(np.arange(5, 30, 5)),
        'min_samples_split': [2, 5, 10],
        'min_samples_leaf': [1, 2, 4],
        'max_features': ['sqrt', 'log2', None]
    }
elif 'SVR' in best_model_type:
    model_to_tune = SVR()
    param_dist = {
        'C': [0.1, 1, 10, 100],
        'kernel': ['linear', 'poly', 'rbf', 'sigmoid'],
        'gamma': ['scale', 'auto', 0.1, 0.01, 0.001],
        'epsilon': [0.01, 0.1, 0.2]
    }
else:
    model_to_tune = Ridge(random_state=42)
    param_dist = {
        'alpha': [0.001, 0.01, 0.1, 1.0, 10.0, 100.0],
        'solver': ['auto', 'svd', 'cholesky', 'lsqr', 'sparse_cg', 'sag', 'saga']
    }

random_search = RandomizedSearchCV(
    estimator=model_to_tune,
    param_distributions=param_dist,
    n_iter=30,
    cv=5,
    scoring='neg_root_mean_squared_error',
    verbose=1,
    random_state=42,
    n_jobs=-1
)

print("Starting hyperparameter tuning with RandomizedSearchCV...")
random_search.fit(X_train_scaled, y_train)

print("\nBest parameters:")
print(random_search.best_params_)

best_model = random_search.best_estimator_

best_y_pred = best_model.predict(X_test_scaled)
best_rmse = calculate_rmse(y_test, best_y_pred)
best_r2 = r2_score(y_test, best_y_pred)

print(f"\nBest Model Performance:")
print(f"RMSE: {best_rmse:.4f}")
print(f"R²: {best_r2:.4f}")
```

```python
from sklearn.ensemble import StackingRegressor
from sklearn.linear_model import Ridge

top_models = results_df.head(5)['model'].tolist()
estimators = [(f"model_{i}", model) for i, model in enumerate(top_models)]

stack_reg = StackingRegressor(
    estimators=estimators,
    final_estimator=Ridge(alpha=0.5),
    n_jobs=-1,
    verbose=1
)

stack_reg.fit(X_train_scaled, y_train)

stack_y_pred = stack_reg.predict(X_test_scaled)
stack_rmse = calculate_rmse(y_test, stack_y_pred)
stack_r2 = r2_score(y_test, stack_y_pred)

print("Stacking Model Performance:")
print(f"RMSE: {stack_rmse:.4f}")
print(f"R²: {stack_r2:.4f}")
```

## Final Model Selection

Let's select the absolute best model between our best single model and the stacked model.

```python
if stack_rmse < best_rmse:
    final_model = stack_reg
    final_rmse = stack_rmse
    final_r2 = stack_r2
    final_model_name = "Stacked Model"
else:
    final_model = best_model
    final_rmse = best_rmse
    final_r2 = best_r2
    final_model_name = f"Tuned {best_model_type}"

print(f"Final Selected Model: {final_model_name}")
print(f"Final RMSE: {final_rmse:.4f}")
print(f"Final R²: {final_r2:.4f}")
```
